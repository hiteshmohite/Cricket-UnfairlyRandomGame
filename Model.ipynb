```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df_dataset = pd.read_csv('Main_Dataset.csv')
list_of_cols = df_dataset.columns.tolist()
list_of_cols = list_of_cols[1:]
df_dataset = df_dataset[list_of_cols]
df_dataset
```

```python
df_modelset = df_dataset.copy()

df_modelset['Feature_1'] = 1
df_modelset['Feature_2'] = 2
df_modelset['Feature_4'] = np.where(df_dataset['Feature_4'] == df_dataset['Feature_1'], 1, 2)
Map = {'bat' : 1, 'field' : 2}
df_modelset['Feature_5'] = df_modelset['Feature_5'].map(Map)
df_modelset['Label'] = np.where(df_dataset['Label'] == df_dataset['Feature_1'], 1, 2)

df_modelset
```

```python
list_of_cols = df_modelset.columns.tolist()
list_of_cols = list_of_cols[: -1]
#list_of_cols = list_of_cols[5:7]
#list_of_cols = list_of_cols[11:]
X = df_modelset[list_of_cols]
y = df_modelset['Label']
#X = df_dataset[list_of_cols]
#y = df_dataset['Label']
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
#logistic regression classifier
clf = LogisticRegression(C=10).fit(X_train, y_train)
```

```python
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
#Knn classifier
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_train, y_train))
print(knn.score(X_test, y_test))
```
