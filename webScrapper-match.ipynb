```python
import numpy as np
import pandas as pd
```

```python
#import urllib2 as link
from urllib.request import urlopen
from bs4 import BeautifulSoup

links = []
team_name = []
#country = []
score = []
overs = []
rpo = []
inns = []
result = []
opposition = []
ground = []
date = []

template_1 = "http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;orderby=team_score;page="
template_2 = ";result=1;result=2;result=3;size=200;template=results;type=team;view=innings"
page_number = 39

for i in range(page_number):
    link = template_1 + str(i+1) + template_2
    links.append(link)
```

```python
df_match_details = pd.DataFrame(columns=['team', 'score', 'overs', 'rpo', 'inns', 'result', 'opposition', 'ground', 'date'])
```

```python
for link in links:

    page = urlopen(link)
    soup = BeautifulSoup(page)
    all_tables=soup.find_all('table')
    right_table=soup.find_all('table', class_='engineTable')
    temp = right_table
    page_core_data = temp[2]

    tag_th = page_core_data.find_all('th')
    
    feature_list_team = []
    for name in tag_th:
        if(name.text):
            feature_list_team.append(name.text)
    #print(feature_list_bowling)
    core_details = page_core_data.find_all('tbody')
    core_details = core_details[0]
    each_page = core_details.find_all('tr', class_='data1')
    temp = each_page[0]
    tag_td = temp.find_all('td')
    for each_row in each_page:
        temp = each_row
        tag_td = temp.find_all('td')
    
        team_name.append(tag_td[0].text)
        score.append(tag_td[1].text)
        overs.append(tag_td[2].text)
        rpo.append(tag_td[3].text)
        inns.append(tag_td[4].text)
        result.append(tag_td[5].text)
        opposition.append(tag_td[7].text)
        ground.append(tag_td[8].text)
        date.append(tag_td[9].text)
```

```python
df_match_details['team'] = team_name
df_match_details['score'] = score
df_match_details['overs'] = overs
df_match_details['rpo'] = rpo
df_match_details['inns'] = inns
df_match_details['result'] = result
df_match_details['opposition'] = opposition
df_match_details['ground'] = ground
df_match_details['date'] = date

df_match_details.to_csv('every_match_details.csv', sep=',')
```
