```python
import numpy as np
import pandas as pd
```

```python
#import urllib2 as link
from urllib.request import urlopen
from bs4 import BeautifulSoup

links = []
batsman_name = []
country = []
runs = []
mins = []
balls_faced = []
fours = []
sixes = []
strike_rate = []
inns = []
start_date = []
against = []
ground = []
    
template_1 = "http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;filter=advanced;orderby=batted_score;page="
template_2 = ";size=200;template=results;type=batting;view=innings"
page_numbers = 443
start_page = 1

for i in range(start_page, page_numbers):
    link = template_1 + str(i) + template_2
    links.append(link)
```

```python
df_batsman_details = pd.DataFrame(columns=['batsman', 'runs', 'mins', 'bf', '4s', '6s', 'SR', 'inns', 'opposition' 'venue', 'Date'])
```

```python
#template_1 = "http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;page="
#template_2 = ";template=results;type=batting;view=innings"
links = []
start_page = 260
page_numbers = 443

for i in range(start_page, page_numbers):
    link = template_1 + str(i) + template_2
    links.append(link)
```

```python
for link in links:

    page = urlopen(link)
    soup = BeautifulSoup(page)
    all_tables=soup.find_all('table')
    right_table=soup.find_all('table', class_='engineTable')
    temp = right_table
    page_core_data = temp[2]

    tag_th = page_core_data.find_all('th')
    #for things in tag_th:
    #    tag_a = things.find_all('a')
        #if len(tag_a) > 0:
        #    inner_tag= tag_a[0]
    #tag_th_left = page_core_data.find_all('th', class_='left')
    #tag_a_black_link = page_core_data.find_all('a', class_='black-link')
    feature_list_batting = []
    for name in tag_th:
        if(name.text):
            feature_list_batting.append(name.text)
    #for name in tag_a_black_link:
    #    feature_list_batting.append(name.text)
    #tag_th_left = page_core_data.find_all('th', class_='left')
    #tag_th_left = tag_th_left[1:]
    #for name in tag_th_left:
    #    feature_list_batting.append(name.text)
    core_details = page_core_data.find_all('tbody')
    core_details = core_details[0]
    each_page = core_details.find_all('tr', class_='data1')
    #temp = each_page[0]
    #tag_td = temp.find_all('td')
    for each_row in each_page:
        temp = each_row
        tag_td = temp.find_all('td')
    
        batsman_name.append(tag_td[0].text)
        runs.append(tag_td[1].text)
        mins.append(tag_td[2].text)
        balls_faced.append(tag_td[3].text)
        fours.append(tag_td[4].text)
        sixes.append(tag_td[5].text)
        strike_rate.append(tag_td[6].text)
        inns.append(tag_td[7].text)
        against.append(tag_td[9].text)
        ground.append(tag_td[10].text)
        start_date.append(tag_td[11].text)
```
